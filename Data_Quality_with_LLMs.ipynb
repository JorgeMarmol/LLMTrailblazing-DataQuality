## 2. Init the model

```python
from vllm import LLM, SamplingParams

path_to_model = (
    "NECOUDBFM/Jellyfish-7B"
)

model = LLM(model=path_to_model)

```

```python
sampling_params = SamplingParams(
    temperature=0.35,
    top_p=0.9,
    max_tokens=1024,
    stop=["### Instruction:"],
)

system_message = "You are an AI assistant that follows instruction extremely well. Help as much as you can."

# You need to define the user_message variable based on the task and the data you want to test on.
user_message = "Tell me ONLY the word HOLA"

prompt = f"{system_message}\n\n### Instruction:\n\n{user_message}\n\n### Response:\n\n"
outputs = model.generate(prompt, sampling_params)
response = outputs[0].outputs[0].text.strip()

print(response)
```

##3.Read Dataset

The dataset used is : https://snap.stanford.edu/data/web-Amazon-links.html

For this task, we will use pandas.

```python
import pandas as pd

def process_review_file(file_path):
    data = []

    current_record = {}

    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.strip()

            if not line:
                if current_record:
                    data.append(current_record)
                    current_record = {}
                continue

            if ": " in line:
                key, value = line.split(": ", 1)
                current_record[key] = value

        if current_record:
            data.append(current_record)

    df = pd.DataFrame(data)

    return df
```

```python
df= process_review_file('Software.txt')[["review/text"]][:100]
```

##4. Sensible Data Detection

Given a tuple and an attribute, our task is to detect whether there is sensitive data in the cell value of this attribute. Sensitive data may include, but are not limited to, personal information, financial details, or confidential information that should be protected.

```python
import re


def sensible_detection(df, column_name):

    total_records= len(df)
    sensible_results = []

    for index, row in df.iterrows():

            attribute_X_value = row[column_name]

            prompt =f"""Your task is to analyze the given text and determine whether it contains any sensitive information. Sensitive data may include personal information, financial details, or confidential content.
Text: [{attribute_X_value}]
Question: Does the text contain any sensitive information?
Your response in a separate line with and ONLY with your final answer. Choose your final answer from [Yes, No]."""

            sampling_params = SamplingParams(
    temperature=0.15,
    top_p=0.9,
    max_tokens=1024,
    stop=["### Instruction:"],
)
            outputs = model.generate(prompt, sampling_params)
            response = outputs[0].outputs[0].text.strip()
            response=re.sub(r'[^\w\s]', '', response)

            sensible_results.append(response)
    df['Sensible Y/N'] = sensible_results
    return df
```

```python
df= sensible_detection(df, "review/text")
```

##5. Sentiment Analysis

Sentiment analysis is the process of using natural language processing (NLP) to determine the emotional tone or opinion expressed in a piece of text. It classifies the sentiment as positive, negative, or neutral, helping to understand the writer's attitude.

```python
def sentiment_analysis(df, column_name):

    total_records= len(df)
    sentiment_results = []

    for index, row in df.iterrows():

            attribute_X_value = row[column_name]
            prompt = prompt = f"""Your task is to analyze the sentiment expressed in the given text. The text may contain emotions, opinions, or feedback, and your goal is to determine the overall sentiment conveyed by the author.
Text: [{attribute_X_value}]
Question: What is the overall sentiment of the text?
Your response in a separate line with and ONLY with your final answer. Choose your final answer from [Positive, Negative, Neutral]."""
            sampling_params = SamplingParams(
    temperature=0.35,
    top_p=0.9,
    max_tokens=1024,
    stop=["### Instruction:"],
)
            outputs = model.generate(prompt, sampling_params)
            response = outputs[0].outputs[0].text.strip()

            sentiment_results.append(response)
    df['Sentiment Analysis'] = sentiment_results
    return df
```

```python
df= sentiment_analysis(df, "review/text")
```

##6. Data Labeling

Data labeling is the process of tagging or annotating data with meaningful labels to train machine learning models. It helps the model understand input data by providing labeled examples, such as identifying objects in images or categorizing text.

```python
def data_labeling(df, column_name):

    total_records= len(df)
    labeling_results = []

    for index, row in df.iterrows():

            attribute_X_value = row[column_name]
            prompt = f"""Your task is to assign the correct label to the given text based on predefined categories. Review the text and determine the category that best fits its content.
Text: [{attribute_X_value}]
Categories: [Product Quality, Customer Service, Delivery and Logistics, Value for Money, Overall Satisfaction, Others]
Question: Which category does the text belong to?
Your response in a separate line with and ONLY with your final answer. Choose your final answer from the provided categories."""
            sampling_params = SamplingParams(
    temperature=0.35,
    top_p=0.9,
    max_tokens=1024,
    stop=["### Instruction:"],
)
            outputs = model.generate(prompt, sampling_params)
            response = outputs[0].outputs[0].text.strip()

            labeling_results.append(response)

    df['Data Label'] = labeling_results
    return df
```

```python
df= data_labeling(df, "review/text")
```

## 7. Summarize Data

```python
def data_summarize(df, column_name):

    total_records= len(df)
    summarize_results = []

    for index, row in df.iterrows():

            attribute_X_value = row[column_name]
            prompt = f"""Your task is to extract the key words from the given text. Review the text and identify the most important words or phrases that represent the main ideas or concepts.
            Text: [{attribute_X_value}]
            Provide the key words separated by commas."""
            sampling_params = SamplingParams(
    temperature=0.35,
    top_p=0.9,
    max_tokens=1024,
    stop=["### Instruction:"],
)
            outputs = model.generate(prompt, sampling_params)
            response = outputs[0].outputs[0].text.strip()

            summarize_results.append(response)

    df['Summarize'] = summarize_results
    return df
```

##8. Extract Structured Data from Unstructured

For text, tools like Named Entity Recognition (NER) or text classification can extract structured information like names, dates, or locations.

The inputs examples have been extracted from: https://medium.com/@ingridwickstevens/extract-structured-data-from-unstructured-text-using-llms-71502addf52b


```python
in1 = """Sweet Delights Bakery introduced lavender-infused vanilla cupcakes with a honey buttercream frosting, using the "Frosting-Spreader-3000". This innovation could inspire our next cupcake creation"""
in2 = """Whisked Away Cupcakes introduced a dessert subscription service, ensuring regular customers receive fresh batches of various sweets. Exploring a similar subscription model using the "SweetSubs" program could boost customer loyalty."""
in3 = """At Velvet Frosting Cupcakes, our team learned about the unveiling of a seasonal pastry menu that changes monthly. Introducing a rotating seasonal menu at our bakery using the "SeasonalJoy" subscription platform and adding a special touch to our cookies with the "FloralStamp" cookie stamper could keep our offerings fresh and exciting for customers."""

inputs = [in1, in2, in3]
results=[]
for input in inputs:

            prompt = f"""Your task is to extract the correct info from unestracted data (text). Return me the data in the format, field:value with string type
Text: [{input}]
schema :  company,offering,advantage, products_and_services, additional_details
"""
            sampling_params = SamplingParams(
    temperature=0.35,
    top_p=0.9,
    max_tokens=1024,
    stop=["### Instruction:"],
)
            outputs = model.generate(prompt, sampling_params)
            response = outputs[0].outputs[0].text.strip()

            results.append(response)
print(results)
```

```python
def string_to_dataframe(string_list):
       data = []
       for string in string_list:
           record = {}
           for line in string.split('\n'):
               key, value = line.split(': ', 1)
               record[key] = value
           data.append(record)
       return pd.DataFrame(data)
```

```python
df_structured= string_to_dataframe(results)
```
